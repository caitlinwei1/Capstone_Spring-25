# Extract and merge vertically of text column on one states for 3 hour interval

```python
from dask.distributed import Client
import dask.dataframe as dd
import pandas as pd


client = Client(
    n_workers=32,          
    threads_per_worker=1,  
    memory_limit='64GB'   
)


target_states =  ['WI']
df = dd.read_parquet(
    "hf://datasets/nyu-dice-lab/wavepulse-radio-raw-transcripts/**/*.parquet",
    columns=['station', 'datetime', 'text', 'state'],
    engine="pyarrow",
    filters=[('state', 'in', target_states)],  
    chunksize="100MB"       
).drop('state', axis=1)     


df['range_key'] = df['datetime'].dt.floor('3H')
df['text'] = df['text'].fillna('')


df = df.shuffle(
    on='station',           
    npartitions=200,        
    shuffle='disk'          
)


grouped = df.groupby(['station', 'range_key'], sort=False).agg({'text': 'sum'})


final_data = grouped.compute().reset_index()
```

```python

output_path = "./WI.parquet"
final_data.to_parquet(
    output_path,
    engine="pyarrow",
    compression="snappy",  
    index=False            
)
```
