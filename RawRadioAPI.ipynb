```python
import requests
import pandas as pd

# Define the API endpoint and parameters
url = "https://datasets-server.huggingface.co/rows"
params = {
    "dataset": "nyu-dice-lab/wavepulse-radio-raw-transcripts",
    "config": "default",
    "split": "train",
    "offset": 0,
    "length": 100  # Fetch 100 rows per request
}

# Initialize an empty list to store all rows
all_rows = []

# Loop through the dataset in chunks
while True:
    # Make the GET request to the API
    response = requests.get(url, params=params)
    
    # Check if the request was successful
    if response.status_code != 200:
        print(f"Failed to fetch data: {response.status_code}")
        break
    
    # Parse the JSON response
    data = response.json()
    
    # Extract the rows from the response
    rows = data['rows']
    
    # Append the rows to the list
    all_rows.extend([row['row'] for row in rows])
    
    # Check if we've reached the end of the dataset
    if len(rows) < params["length"]:
        print("Reached the end of the dataset.")
        break
    
    # Update the offset for the next request
    params["offset"] += params["length"]
    print(f"Fetched {params['offset']} rows so far...")

# Convert the list of rows to a Pandas DataFrame
df = pd.DataFrame(all_rows)

# Display the first few rows of the DataFrame
print(df.head())
print(f"Total rows in the dataset: {len(df)}")
```
